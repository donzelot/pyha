```python

import matplotlib.pyplot as plt
%matplotlib notebook
import numpy as np

# 2hz signal
t = np.linspace(0,2, 512)
a = np.exp(1j * 2*np.pi*2 * t)

# 4hz signal
b = np.exp(1j * 2*np.pi*4 * t)
```

```python
plt.plot(a)
plt.plot(b)
plt.show()
```

```python
res = a*b
print(res)
plt.plot(a*b)
plt.show()
```
