```python
from common.sfix import Sfix

class Counter():
    def __init__(self, step):
        self.step = step
        self.counter = 0.0

    def run(self):
        fmin = -1
        fmax = +1
        val = self.counter + self.step
        
        res = (val-fmin) % (fmax - fmin) + fmin
        print('V: {}, W: {}'.format(val, res))
        self.counter = res
        return self.counter
```

```python
dut = Counter(0.01)

res = []
for _ in range(210):
    res.append(dut.run())

plt.plot(res)
```
