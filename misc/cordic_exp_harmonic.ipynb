```python
%matplotlib inline
import numpy as np
import scipy
from scipy.signal import hilbert, chirp, spectrogram, periodogram
import matplotlib.pyplot as plt
from hwpy.common.cordic import CORDIC

fs = 1e3
periods = 2
freq = 1
```

```python

phase_inc = 2 * np.pi * freq / fs
phase_list = np.arange(0, periods * fs * phase_inc, phase_inc)

phase = np.sin(phase_list)
plt.plot(phase)
plt.title('phase')
plt.show()
# plt.plot(phase)
# plt.show()

ref = np.exp(phase * 1j)
plt.plot(ref.real)
plt.plot(ref.imag)
plt.title('Numpy exp')
plt.show()

cord = CORDIC(18)
res = cord.exp(phase * 1j)

plt.plot([x.real for x in res])
plt.plot([x.imag for x in res])
plt.title('CORDIC exp')
plt.show()
```

# Compare

```python
np.testing.assert_almost_equal(res, ref, decimal=5)
```
